```python
import pandas as pd

path="/content/drive/MyDrive/sample_sales_data.csv"
df = pd.read_csv(path)
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Plot distribution of a numerical column
sns.histplot(df['quantity'], kde=True)
plt.title('Distribution of quantity Column')
plt.show()
```

```python
sns.pairplot(df[['unit_price', 'payment_type', 'total']])
plt.suptitle('Pairwise Scatter Plot')
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Count plot for a categorical column
sns.countplot(x='category', data=df)
plt.title('Count Plot for category Column')
plt.show()
```

```python
sns.boxplot(x='customer_type', y='unit_price', data=df)
plt.title('Box Plot for customer_type vs. unit_price')
plt.show()
```
